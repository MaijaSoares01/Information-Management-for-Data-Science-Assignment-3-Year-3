**What is Clustering?**<br>
Clustering is unsupervised machine learning. This means there is one set of input data that is not labelled, we extract information from this input data but we do not know what the output will be. Clustering points out similarities between objects and things by which they can be grouped by, groups with overlapping characteristics cluster while groups with different characteristics do not cluster.

<br>**Examples of Clustering**<br>
On streaming services such as YouTube, viewers of a certain type of content can have overlapping interest with those who watch another type of content therefore that content can be advertised to them. If it is shown that they may not like a certain type of content based on what they watch, those videos will be advertised less or not at all.

<br>**What is Classification?**<br>
Clustering is supervised machine learning. This means there is one set of input that is labelled, we extract information from this input data and we know what the output will be from the algorithm. Classification means that objects and things are assigned into predefined classes.<br>
*There are 2 types of Classification:*

<li>Multiclassification-More than two classes can be predefined and objects and things can be assigned to them. For example, fruits like apple, orange, banana and kiwi</li>
<li>Binary Classification-Only two classes are needed are predefined and objects and things can be assigned to them. For example, if a message is spam or not spam.</li>
</ul>

<br>**Examples of Classification**<br>
The classification of different types of animals and what category and family an individual animal belongs to classifies it. For example the most common classification types for animals are mammals, fish, birds, reptiles, and amphibians, this is Multiclassification as there are more than 2 classes.

**What is the Map Function?**<br>
A high order function that accepts another function and a list or sequence of iterables as parameters and returns an output, once each iterable has run through the function. Example, map(function, iterables)
<br>**Benefits of the Map Function**<br>
Without using a explicit loop all iterables in the list will run through the funtion. Mapping allows a list of iterables to be transformed and processed  into a new list of iterables.

<br>**What is the Reduce Function?**<br>
The reduce function once imported merges the list of iterables into one iterable(adds up all iterables).Example, reduce(function, iterables)
<br>**Benefits of the Reduce Function**<br>
Without using a explicit loop all iterables in the list will run through the funtion and combines all iterables into one. A large list of iterables can be compressed into one iterable. 

<br>**What is the Filter Function?**<br>
The filter funtion returns the wanted iterables by filtering through the list into a new list. Example, filter(function, iterables)
<br>**Benefits of the Filter Function**<br>
Without using a explicit loop all iterables in the list will run through the funtion, all iterables will be filtered. This is to 'get rid' of unwanted iterables.

image.png

```python
#Map example without lamda
def f(x):
    return x*2
x = map(f, (1,2,3,4))#will display last, first, second, third iterable
print(set(x))#use print set to display returns
x = tuple(map(f, (1,2,3,4)))#tuple will display iterables in order
print(x)#use print to display returns 
#Map example with lamda
numbers = [1,2,3,4]
new_numbers = map(lambda x: x*2 , numbers)#will display last, first, second, third iterable no call to function, function is within call
print(set(new_numbers))#use print set to display returns
new_numbers = tuple(map(lambda x: x*2 , numbers))#tuple will display iterables in order no call to function, function is within call
print(new_numbers)#will display iterables in orderd
```

**Overfitting Vs Underfitting**<br>
Overfitting takes place when a machine learning algorithm or statistical model takes into account data that is just noise or not necessary. It also occurs when the algorithm or model fits the data far too well, shows low bias but high variance. Overfitting relies heavily on th training data. Another characterisation is good performance on the training data and poor generliazation to other data. Put simply a overfitting algorithm or model is an excessively complicated one. This can be shown below where the overfitting diagrams take into account data that are not needed and over complicate the model, shown by the scribbled trend path.<br> 

Where as Underfitting takes place when a machine learning algorithm or statistical model does not take into account the primary trend(important data points). It also occurs when the algorithm or model does not fit the data, shows high bias but low variance. Underfitting fails to learn the relationships in the training data which is why it doesnt represent the underlying trend. Another characterisation is poor performance on the training data and poor generliazation to other data. Put simply a underfitting algorithm or model is an excessively simple one. This can be shown below where the underfitting diagrams do not take into account data  that show the underlying trends and therfore over simplify the model, shown by the trend path that does not follow the data points.<br>

Both overfitting and underfitting lead to poor predictions on new data sets.

<br>**Approach in Relation to Overfitting and Underfitting**<br>
*How to Handle Underfitting model or algorithm*<br>
*   Get more training data.
*   Increase the complexity of the model.
*   Increasing the training time, until cost function is minimised.
*   Increase the size or number of parameters in the model.

<br>*How to Handle Overfitting model or algorithm*<br>
*   Remove layers to reduce network capacity or reduce number of elements in hidden layers.
*   Randomly remove features by setting the dropout layers to zero.
*   Using validation and cross validation to fit more than one model to compare their predictive accuracies on test data.
*   By adding cost to the loss function for large weights, this is called regularization.

The Nearest Centroid classifier works on a simple principle : Given a data point (observation), the Nearest Centroid classifier simply assign it the label (class) of the training sample whose mean or centroid is closest to it. 
How the nearest centroid classifier works? 
Basically, what the nearest centroid classifier does can be explained in three steps: 
 

The centroid for each target class is computed while training.
After training, given any point, say ‘X’. The distances between the point X and each class’ centroid is calculated.
Out of all the calculated distances, the minimum distance is picked. The centroid to which the given point’s distance is minimum, it’s class is assigned to the given point.
